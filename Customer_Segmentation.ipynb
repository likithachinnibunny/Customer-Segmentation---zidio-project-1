**Feature scaling**

```python
#Feature scaling
x=df.iloc[:,2: ].values
x
```

```python
scale = MinMaxScaler()
features = scale.fit_transform(x)
features
```

**Elbow Method for Optimal Clusters**

```python
#Elbow method for optimal clusters
inertia = []
for i in range(1, 16):
    k_means = KMeans(n_clusters=i, n_init=10)
    k_means = k_means.fit(features)
    inertia.append(k_means.inertia_)

print(inertia)
```

```python
# elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')
plt.xlabel('No of clusters')
plt.ylabel('Inertia')

# Kelbow visualizer
plt.subplot(1,2,2)
kmeans = KMeans(n_init=10)
visualize = KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
```

**Silhoutte Analysis for Optimal Clusters**

```python
#Silhoutte Analysis for optimal clusters
silhouette_avg = []
for i in range(2,16):
    # initialise kmeans
    kmeans = KMeans(n_clusters=i,n_init=10)
    cluster_labels = kmeans.fit_predict(features)
    # silhouette score
    silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

**K-Means Clustering**

```python
#K-means clustering
model = KMeans(n_clusters=3,n_init=10)
model = model.fit(features)
```

```python
y_km = model.predict(features)
centers = model.cluster_centers_
```

```python
df['Cluster'] = pd.DataFrame(y_km)
df.to_csv("Cluster_data",index=False)
```
